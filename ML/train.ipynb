```python
#menampilkan dataset
import os

aloperia_dir = os.path.join('D:\capstone_Xplisit/ML/nail diseases/train/aloperia areata')
beaus_dir = os.path.join('D:\capstone_Xplisit/ML/nail diseases/train/beaus lines')
bluish_dir = os.path.join('D:\capstone_Xplisit/ML/nail diseases/train/bluish nail')

print('total training aloperia areatas:', len(os.listdir(aloperia_dir)))
print('total training beaus images:', len(os.listdir(beaus_dir)))
print('total training bluish images:', len(os.listdir(bluish_dir)))

aloperia_files = os.listdir(aloperia_dir)
print(aloperia_files[:3])
beaus_files = os.listdir(beaus_dir)
print(beaus_files[:3])
bluish_files = os.listdir(bluish_dir)
print(bluish_files[:3])
```

```python
#show dataset
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

pic_index = 4

next_aloperia = [os.path.join(aloperia_dir, fname) 
                for fname in aloperia_files[pic_index-2:pic_index]]
next_beaus = [os.path.join(beaus_dir, fname) 
                for fname in beaus_files[pic_index-2:pic_index]]
next_bluish = [os.path.join(bluish_dir, fname) 
                for fname in bluish_files[pic_index-2:pic_index]]

for i, img_path in enumerate(next_aloperia+next_beaus+next_bluish):
  img = mpimg.imread(img_path)
  plt.imshow(img)
  plt.axis('Off')
  plt.show()
```
